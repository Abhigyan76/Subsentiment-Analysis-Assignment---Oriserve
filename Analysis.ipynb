```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
```

```python
metric = open('bert_base_metrics.pkl', 'rb') 
```

```python
k = pickle.load(metric) 
metric.close() 
```

```python
df = pd.DataFrame(k)
```

#### Training and Validation BCE-Loss Plot

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(x='epochs', y='train_loss', data=df, color='red', label='Training Loss')
sns.lineplot(x='epochs', y='val_loss', data=df, color='blue', label='Validation Loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')

plt.legend()
plt.show()
```

###### As I can se at epoch 5 train and validation are low and after that validaton loss is greater as compared to training, So there is overfitting after 5th Epoch

#### Training and Validation F1-Micro-Score Plot

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(x='epochs', y='training_f1_micro', data=df, color='red', label='Training F1-Micro-Score')
sns.lineplot(x='epochs', y='val_f1_micro', data=df, color='blue', label='Validation F1-Micro-Score')

plt.xlabel('Epochs')
plt.ylabel('F1-Micro-Score')
plt.title('Training and Validation F1-Micro-Score')

plt.legend()
plt.show()
```

###### Here also I can see that after 5th epoch validation F1-Micro-Score is not increasing while training F1-Micro-Score is increasing which shows overfitting after 5th Epoch, Training F1-Micro-Score reached to .90 at 10th Epoch

#### Training and Validation F1-Macro-Score Plot

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(x='epochs', y='training_f1_macro', data=df, color='red', label='Training F1-Macro-Score')
sns.lineplot(x='epochs', y='val_f1_macro', data=df, color='blue', label='Validation F1-Macro-Score')

plt.xlabel('Epochs')
plt.ylabel('F1-Macro-Score')
plt.title('Training and Validation F1-Macro-Score')

plt.legend()
plt.show()
```

###### Here also I can see that after 5th epoch validation F1-Macro-Score is not increasing while training F1-Macro-Score is increasing which shows overfitting after 5th Epoch, Training F1-Macro-Score reached to .85 at 10th Epoch

#### Training and Validation Hamming Loss Plot

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(x='epochs', y='training_hamming_loss', data=df, color='red', label='Training Hamming-Loss')
sns.lineplot(x='epochs', y='val_hamming_loss', data=df, color='blue', label='Validation Hamming-Loss')

plt.xlabel('Epochs')
plt.ylabel('Hamming-Loss')
plt.title('Training and Validation Hamming-Loss')

plt.legend()
plt.show()
```

###### As I can se at epoch 5 train and validation are low and after that validaton loss is greater as compared to training, So there is overfitting after 5th Epoch
